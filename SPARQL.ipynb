```python
from rdflib import Graph

# Load RDF data from the Turtle file
g = Graph()
g.parse("Ourcommon.ttl", format="ttl")

# Define SPARQL queries
queries = [
    # SPARQL Query 1: Retrieve all roles associated with an organization country
    """
    PREFIX ex: <http://example.org/>
    PREFIX org: <http://example.org/org/>
    SELECT ?role
    WHERE {
        <http://example.org/organization_country> ex:organization_country_role ?role .
    }
    """,
    # SPARQL Query 2: Retrieve all resources associated with a specific resource country
    """
    PREFIX ex: <http://example.org/>
    PREFIX org: <http://example.org/org/>
    SELECT ?resource
    WHERE {
        ?resource ex:org_resource_resource_country <http://example.org/resource_country> .
    }
    """,
    # SPARQL Query 3: Retrieve roles associated with both groups and resources
    """
    PREFIX ex: <http://example.org/>
    PREFIX org: <http://example.org/org/>
    SELECT ?role
    WHERE {
        ?group ex:org_group_role ?role .
        ?resource ex:org_resource_role ?role .
    }
    """,
    # SPARQL Query 4: Retrieve all groups associated with a specific organization country
    """
    PREFIX ex: <http://example.org/>
    PREFIX org: <http://example.org/org/>
    SELECT ?group
    WHERE {
        ?group ex:org_group_organization_country <http://example.org/organization_country> .
    }
    """,
    # SPARQL Query 5: Retrieve organization country linked to both groups and resources
    """
    PREFIX ex: <http://example.org/>
    PREFIX org: <http://example.org/org/>
    SELECT ?organizationCountry
    WHERE {
        ?group ex:org_group_organization_country ?organizationCountry .
        ?resource ex:org_resource_organization_country ?organizationCountry .
    }
    """
]

# Execute each query and print results
for i, query in enumerate(queries, 1):
    print(f"Executing Query {i}:")
    results = g.query(query)
    for row in results:
        # Print the results in the desired format
        print(f"ex:{row[0].split('/')[-1]}")
    print("\n")
```

```python
from rdflib import Graph

# Load RDF data from the Turtle file
g = Graph()
g.parse("Ourcommon.ttl", format="ttl")

# SPARQL query to retrieve all predicates with the prefix ex:
query = """
PREFIX ex: <http://example.org/>
SELECT DISTINCT ?predicate
WHERE {
    ?subject ?predicate ?object .
    FILTER(STRSTARTS(STR(?predicate), "http://example.org/"))
}
"""

# Execute the query
results = g.query(query)

# Print results in the desired format
print("All ex predicates:")
for row in results:
    predicate = row[0]
    # Extract and format the predicate to match the desired output
    ex_predicate = predicate.split('/')[-1]
    print(f"ex:{ex_predicate} .")
```
